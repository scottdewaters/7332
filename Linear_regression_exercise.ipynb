```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
#data prep
data_root = "https://github.com/ageron/data/raw/main/"
lifesat = pd.read_csv(data_root + "lifesat/lifesat.csv")
x = lifesat[["GDP per capita (USD)"]].values
y = lifesat[["Life satisfaction"]].values
```

```python
#visualize
lifesat.plot(kind='scatter', grid=True,
             x="GDP per capita (USD)",
             y="Life satisfaction")
plt.axis([23_500, 62_500, 4, 9])
plt.show()
```

```python
#model selection
model = LinearRegression()
```

```python
#train model
model.fit(x, y)
```

```python
#Predict for new country
X_new = [[37_665.2]] #Cyprus' GDP per capita in 2020
print(model.predict(X_new))
```

```python
# prompt: Visualize the model on a plot chart with X_new as a different color
# Predict for new country
X_new = [[37_665.2]] # Cyprus' GDP per capita in 2020
y_pred = model.predict(X_new)
print(y_pred)

# Visualize the model and the new prediction
lifesat.plot(kind='scatter', grid=True,
             x="GDP per capita (USD)",
             y="Life satisfaction")
plt.axis([23_500, 62_500, 4, 9])

# Plot the regression line
X_plot = np.linspace(23_500, 62_500, 100).reshape(-1, 1)
y_plot = model.predict(X_plot)
plt.plot(X_plot, y_plot, color='red')

# Plot the new data point
plt.scatter(X_new, y_pred, color='green', marker='*', s=200, label='Cyprus')
plt.legend()
plt.show()
```
